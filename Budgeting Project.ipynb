```python
class Budgeting:
    def __init__(self):
        self.income = 0
        self.expenses = 0
        self.expense_list = []
        self.expense_name = []
        self.income_name = []
        self.income_list = []
        self.prompt_income()

    def income_ask(self):
        add_income = input('Add income? [yes or no]: ')
        return add_income

    def income_sum(self):
        self.income = sum(self.income_list)

    def expense_ask(self):
        add_expense = input('Add expense? [yes or no]: ')
        return add_expense

    def expense_sum(self):
        self.expenses = sum(self.expense_list)

    def prompt_income(self):
        while True:
            result = self.income_ask()
            if result == 'yes':
                try:
                    income_input = float(input("Enter amount of income: "))
                    self.income_list.append(income_input)
                    income_name = input("Enter name of income: ")
                    self.income_name.append(income_name)
                except ValueError:
                    print("Invalid input. Please enter a valid number.")
            else:
                self.income_sum()
                if not self.income_list:
                    print("Please enter at least one source of income.")
                    self.prompt_income()
                else:
                    name = [name for name in self.income_name]
                    income = [income for income in self.income_list]
                    income_dict = dict(zip(name, income))
                    for i in income_dict:
                        print(i + ':', '$' + str(income_dict[i]))
                    print("You have income of:", '$' + str(self.income))
                    self.prompt_expense()
                break

    def prompt_expense(self):
        while True:
            result = self.expense_ask()
            if result == 'yes':
                try:
                    expense_input = float(input("Enter amount of expense: "))  # Use float to handle decimal values
                    self.expense_list.append(expense_input)
                    expense_name = input("Enter name of expense: ")
                    self.expense_name.append(expense_name)
                except ValueError:
                    print("Invalid input. Please enter a valid number.")
            else:
                self.expense_sum()
                if not self.expense_list:
                    print("Please enter at least one expense.")
                    self.prompt_expense()
                else:
                    name = [name for name in self.expense_name]
                    expense = [expense for expense in self.expense_list]
                    expense_dict = dict(zip(name, expense))
                    for i in expense_dict:
                        print(i + ':', '$' + str(expense_dict[i]))
                    print("You have expenses of:", '$' + str(self.expenses))
                    self.value()
                break

    def value(self):
        value_put = self.income - self.expenses
        if value_put > 0:
            print("You have a surplus of:", '$' + str(value_put))
            self.budget(value_put)
        elif value_put == 0:
            print("You broke even")
        else:
            print("You are at a deficit of:", '$' + str(value_put))
            self.budget(value_put)

    def budget(self, val_put):
        x = val_put
        if x > 0:
            while True:
                print("Select number")
                print("1: Stock Investment")
                print("2: Bond Investment")
                print("3: Vacation Fund")
                print("4: Savings")
                print("5: Exit")
                choice = int(input("Enter your choice (1-5):"))

                if choice in [1, 2, 3, 4, 5]:

                    if choice == 1:
                        investment_percentage = float(input(f"Enter the percentage you would like to invest in your choice: "))
                        invested_amount = val_put * (investment_percentage / 100)
                        print(f"You have invested ${invested_amount} into the stock market.")
                        x = x - invested_amount
                    elif choice == 2:
                        investment_percentage = float(input(f"Enter the percentage you would like to invest in your choice: "))
                        invested_amount = val_put * (investment_percentage / 100)
                        print(f"You have invested ${invested_amount} into the bond market.")
                        x = x - invested_amount
                    elif choice == 3:
                        investment_percentage = float(input(f"Enter the percentage you would like to invest in your choice: "))
                        invested_amount = val_put * (investment_percentage / 100)
                        print(f"You have put ${invested_amount} into your vacation fund.")
                        x = x - invested_amount
                    elif choice == 4:
                        investment_percentage = float(input(f"Enter the percentage you would like to invest in your choice: "))
                        invested_amount = val_put * (investment_percentage / 100)
                        print(f"You have put ${invested_amount} into savings.")
                        x = x - invested_amount
                    if choice == 5:
                        break
                    val_put -= invested_amount
                    print(f"You now have ${val_put} left in income.")
                    
                else:
                    print("Invalid input. Pick a number 1-5.")
                if choice==5:
                    break

        elif x < 0:
            print(f"Withdraw ${val_put} from savings to cover expenses.")

if __name__ == '__main__':
    Budgeting()
```
